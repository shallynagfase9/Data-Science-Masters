```python
import seaborn as sns
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
########### Standardization -> Use z-score formula to normalize data
import numpy as np
mean = np.mean(df["total_bill"])
std = np.std(df["total_bill"])
print(mean,std)
```

```python
normalized_data=[]
for i in list(df['total_bill']):
    z_score=(i-mean)/std
    normalized_data.append(z_score)

```

```python
normalized_data
```

```python
sns.histplot(df["total_bill"])
```

```python
sns.histplot("normalized_data")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```

```python
import pandas as pd
pd.DataFrame(scaler.fit_transform(df[["total_bill","tip"]]),columns=["total_bill","tip"])
```

```python
scaler.transform([[13,4]])
```

```python
#### Normalization - Min Max scaler (The Values are scaled between 0 to 1)
df = sns.load_dataset("taxis")
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()

min_max.fit_transform(df[["distance","fare","tip"]])
```

```python
min_max.transform([[13.4,4.40,5.0]])
```

```python
######### Unit Vactor - Magnitude is 1 ############

from sklearn.preprocessing import normalize
pd.DataFrame(normalize(df[["distance","fare","tip"]]),columns = ["distance","fare","tip"])
```
