```python
df5 = pd.DataFrame({"date":["2022-06-19",'2022-06-22','2022-03-22']})
df5
```

```python
df5.dtypes
```

```python
df5["update_date"] = pd.to_datetime(df5['date'])
df5
```

```python
df5.dtypes
```

```python
df5["year"] = df5["update_date"].dt.year
df5
```

```python
df5["month"] = df5["update_date"].dt.month
df5
```

```python
df5["day"] = df5["update_date"].dt.day
df5
```
