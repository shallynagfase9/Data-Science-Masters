```python
def inr_con(x):
  return x*80

data["Fare_INR"] = data["Fare"].apply(inr_con)

data.head(5)
```

```python
def create_flag(x) :
  if x < 10 :
    return "cheap"
  elif x >= 10 and x < 20 :
    return "mid"
  else :
    return "high"

data["Flag_fare"] = data["Fare"].apply(create_flag)

data.head(2)
```
