```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, c = "red")
plt.title("This is my graph")
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
```

```python
plt.scatter(x,y, c = "#e3117a",alpha = .5, marker = "x")  # alpha will be able to control the intensity of the color
# Marker will specify the dot in the x format
plt.title("This is my graph")
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
```

```python
plt.figure(figsize=(2,6)) # here i can specify the graph size
plt.scatter(x,y, c = "red")
plt.title("This is my graph")
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
```

```python
x = np.linspace(1,10, (100))
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y, c = "b")
```

```python
plt.figure(figsize=(8,2))
plt.scatter(x,y,c='pink')
plt.title("This is my sin wave graph")
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.grid()
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,4))
plt.bar(x,y, color='purple')
plt.title("This is my bar plot")
plt.xlabel("Representing my cat data")
plt.ylabel("Representing my num data")
plt.grid()
```

```python
plt.figure(figsize=(5,4))
plt.barh(x,y, color='black') # barh() function will flip the x and y axis
plt.title("This is my bar plot")
plt.xlabel("Representing my cat data")
plt.ylabel("Representing my num data")
plt.grid()
```

```python
plt.figure(figsize=(4,4))
x = [2,5,3,5,6,1,2]
y = [2,5,3,5,6,1,2]
plt.plot(x,y)
plt.title("This is my graph")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.grid()
```

```python
plt.figure(figsize=(5,4))
plt.scatter(x,y)
```

```python
data = [1,2,3,4,5,6,8,8,8,9]
plt.figure(figsize=(8,4))
plt.hist(data, color="yellow")
plt.title(" Hist plot Example ")
plt.xlabel("Representing unique values")
plt.ylabel("Representing Frequency")
plt.grid()
plt.savefig('test.png')
```

```python
# 3D Graph Example
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(x,y,z, c ="green", alpha=.8)
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.show()
```
