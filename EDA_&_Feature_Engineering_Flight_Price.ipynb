```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Feature Enngineering
df["Date"] = df["Date_of_Journey"].str.split("/").str[0]
```

```python
df["Date_of_Journey"].str.split("/").str[0]
```

```python
df["Date"]
```

```python
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Month"]
```

```python
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]
df["Year"]
```

```python
df.head()
```

```python
df.info()
```

```python
# Changing the type of column ["Date","Month","Year"] from object to int
```

```python
df["Date"] = df["Date"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df['Year'].astype(int)
```

```python
df.info()
```

```python
# For permanent changes in dataset use inplace = "True"
# for column axis = 1
# Drop Date_of_Journey
df.drop("Date_of_Journey", axis=1, inplace=True)
```

```python
df.head()
```

```python
df["Arrival_Time"] = df["Arrival_Time"].apply(lambda x : x.split(" ")[0])
df["Arrival_Time"]
```

```python
df["Arrival_hour"] = df["Arrival_Time"].str.split(":").str[0]
df["Arrival_hour"]
```

```python
df["Arrival_min"] = df["Arrival_Time"].str.split(":").str[1]
df["Arrival_min"]
```

```python
df["Arrival_hour"] = df["Date"].astype(int)
df["Arrival_min"] = df["Month"].astype(int)
```

```python
df.info()
```

```python
df.drop("Arrival_Time",inplace=True, axis=1)
```

```python
df["Dept_Hour"] = df["Dep_Time"].str.split(":").str[0]
df["Dept_min"] = df["Dep_Time"].str.split(":").str[1]
```

```python
df["Dept_Hour"]
```

```python
df["Dept_min"]
```

```python
df.head(2)
```

```python
df["Dept_Hour"] = df["Dept_Hour"].astype(int)
df["Dept_min"] = df["Dept_min"].astype(int)
```

```python
df.info()
```

```python
df.drop("Dep_Time",inplace=True,axis=1)
```
